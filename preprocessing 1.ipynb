```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('movie_metadata.csv')
```

```python
# we have movies only upto 2016
import matplotlib.pyplot as plt
%matplotlib inline
data.title_year.value_counts(dropna=True).sort_index().plot(kind='barh',figsize=(15,16))
plt.show()
```

```python
# recommendation will be based on these features only
data = data.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title','plot_keywords','language']]
```

```python
data.head(10)
```

```python
data['actor_1_name'] = data['actor_1_name'].replace(np.nan, 'unknown_actor')
data['actor_2_name'] = data['actor_2_name'].replace(np.nan, 'unknown_actor')
data['actor_3_name'] = data['actor_3_name'].replace(np.nan, 'unknown_actor')
data['director_name'] = data['director_name'].replace(np.nan, 'unknown_director')
data['plot_keywords'] = data['plot_keywords'].replace(np.nan, 'unknown_keywords')
data['language'] = data['language'].replace(np.nan, 'English')
```

```python
data
```

```python
data['genres'] = data['genres'].str.replace('|', ' ')
```

```python
data['plot_keywords']= data['plot_keywords'].str.replace('|',' ')
```

```python
data
```

```python
for i in data.columns:
    data[i] = data[i].str.lower()
```

```python
# null terminating char at the end
data['movie_title'][1]
```

```python
# removing the null terminating char at the end
data['movie_title'] = data['movie_title'].apply(lambda x : x[:-1])
```

```python
data['movie_title'][1]
```

```python
data.to_csv('data.csv',index=False)
```
