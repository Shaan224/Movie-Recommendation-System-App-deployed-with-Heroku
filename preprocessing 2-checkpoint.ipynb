```python
import numpy as np
import pandas as pd
```

```python
credits = pd.read_csv('credits.csv')
```

```python
keywords = pd.read_csv('keywords.csv')
```

```python
credits
```

```python
meta = pd.read_csv('movies_metadata.csv')
```

```python
meta['release_date'] = pd.to_datetime(meta['release_date'], errors='coerce')
```

```python

meta['year'] = meta['release_date'].dt.year
```

```python
meta['year'].value_counts().sort_index()
```

```python
# Getting only 2017 movies as we already have movies up to the year 2016 in preprocessing 1 file. 
# We don't have enough data for the movies from 2018, 2019 and 2020. 
# We'll deal with it in the upcoming preprocessing files
new_meta = meta.loc[meta.year == 2017,['genres','id','title','year','original_language']]
```

```python

new_meta
```

```python
link = "https://meta.wikimedia.org/wiki/Template:List_of_language_names_ordered_by_code"
df1 = pd.read_html(link, header=0)[0]
df1
```

```python
new_meta = pd.merge(new_meta, df1, left_on='original_language',right_on='Code')
```

```python
new_meta.drop(['original_language','Code','Directionality','Local language name','Local or English Wikipedia article','Comment'],axis=1,inplace=True)
```

```python
new_meta = new_meta.rename(columns={'English language name':'language'})
```

```python
new_meta['id'] = new_meta['id'].astype(int)
```

```python
data = pd.merge(new_meta, credits, on='id')
```

```python
data = pd.merge(data, keywords, on='id')
```

```python
pd.set_option('display.max_colwidth', 75)
data
```

```python
# evaluates an expression node or a string containing a Python literal or container display
import ast
data['genres'] = data['genres'].map(lambda x: ast.literal_eval(x))
data['cast'] = data['cast'].map(lambda x: ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x: ast.literal_eval(x))
data['keywords'] = data['keywords'].map(lambda x: ast.literal_eval(x))
```

```python
def make_genresList(x):
    gen = []
    st = " "
    for i in x:
        if i.get('name') == 'Science Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.NaN
    else:
        return (st.join(gen))
```

```python
data['genres_list'] = data['genres'].map(lambda x: make_genresList(x))
```

```python
data['genres_list']
```

```python
def get_actor1(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == []:
        return np.NaN
    else:
        return (casts[0])
```

```python
data['actor_1_name'] = data['cast'].map(lambda x: get_actor1(x))
```

```python
def get_actor2(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=1:
        return np.NaN
    else:
        return (casts[1])
```

```python
data['actor_2_name'] = data['cast'].map(lambda x: get_actor2(x))
```

```python
data['actor_2_name']
```

```python
def get_actor3(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=2:
        return np.NaN
    else:
        return (casts[2])
```

```python
data['actor_3_name'] = data['cast'].map(lambda x: get_actor3(x))
```

```python
data['actor_3_name']
```

```python
def get_directors(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return (st.join(dt))
```

```python
data['director_name'] = data['crew'].map(lambda x: get_directors(x))
```

```python
data['director_name']
```

```python
def make_keywordList(x):
    gen = []
    st = " "
    for i in x:
        if i.get('name') == 'Science Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.NaN
    else:
        return (st.join(gen))
```

```python
data['keywords'] = data['keywords'].map(lambda x: make_keywordList(x))
```

```python
data['keywords'] 
```

```python
data
```

```python
movie = data.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres_list','title','keywords','language']]
```

```python
movie
```

```python
movie.isna().sum()
```

```python
movie['actor_1_name'] = movie['actor_1_name'].replace(np.nan, 'unknown_actor')
movie['actor_2_name'] = movie['actor_2_name'].replace(np.nan, 'unknown_actor')
movie['actor_3_name'] = movie['actor_3_name'].replace(np.nan, 'unknown_actor')
movie['director_name'] = movie['director_name'].replace(np.nan, 'unknown_director')
movie['keywords'] = movie['keywords'].replace(np.nan, 'unknown_keywords')
movie['language'] = movie['language'].replace(np.nan, 'English')
movie['genres_list'] = movie['genres_list'].replace(np.nan, ' ')
```

```python
movie.isna().sum()
```

```python
movie = movie.rename(columns={'genres_list':'genres'})
movie = movie.rename(columns={'title':'movie_title'})
movie = movie.rename(columns={'keywords':'plot_keywords'})
```

```python
for i in movie.columns:
    movie[i] = movie[i].str.lower()
```

```python
movie['comb'] = movie['actor_1_name'] + ' ' + movie['actor_2_name'] + ' '+ movie['actor_3_name'] + ' '+ movie['director_name'] +' ' + movie['genres']+' ' + movie['plot_keywords']+' ' + movie['language']
```
